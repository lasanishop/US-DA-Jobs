```python
# Using loc with axis=0 to select specific rows
df.loc[2:5, :]  # Select rows 2 to 5 (inclusive) and all columns
df.loc[[1, 3, 5], :]  # Select rows 1, 3, and 5 and all columns

# Using loc with axis=1 to select specific columns
df.loc[:, 'location']  # Select all rows and 'column1'
df.loc[:, ['company', 'salary estimate']]  # Select all rows and columns 'column1' and 'column2'

# Using iloc with axis=0 to select specific rows
df.iloc[2:5, :]  # Select rows 2 to 4 (exclusive) and all columns
df.iloc[[1, 3, 5], :]  # Select rows 1, 3, and 5 and all columns

# Using iloc with axis=1 to select specific columns
df.iloc[:, 0]  # Select all rows and the first column
df.iloc[:, [0, 2]]  # Select all rows and columns 0 and 2
```
